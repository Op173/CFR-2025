## Part 3: Advanced Customization and Subplots (30 minutes)


#### Example 6: Creating Subplots

Subplots allow you to create multiple plots in a single figure.

```python
dir(np)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 11)
line_data = np.random.rand(8, 10)  # 8 sets of line/bar data

# Create the figure with 2 rows and 4 columns
fig, axs = plt.subplots(2, 4, figsize=(20, 10))
axs = axs.flatten()  # Flatten for easy looping

# Plot alternating line and bar charts
for i in range(8):
    if i % 2 == 0:  # Even indices: line plot
        axs[i].plot(x, line_data[i], color='red')
        axs[i].set_title(f'Line Plot {i+1}')
    else:           # Odd indices: bar plot
        axs[i].bar(x, line_data[i], color='green')
        axs[i].set_title(f'Bar Chart {i+1}')
    
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

**Practice Problem 6: Create Your Own Subplots**

Create a figure with two subplots:
- A line plot for the data x: [0, 1, 2, 3, 4], y: [0, 1, 4, 9, 16]
- A bar plot for the data categories: ['A', 'B', 'C'], values: [5, 10, 15]

**Solution:**


```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Data for the line plot
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Data for the bar plot
categories = ['A', 'B', 'C']
values = [5, 10, 15]

# Create subplots: 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Line plot on the first subplot
axs[0].plot(x, y, marker='o', color='blue')
axs[0].set_title('Line Plot')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')

# Bar plot on the second subplot
axs[1].bar(categories, values, color='green')
axs[1].set_title('Bar Plot')
axs[1].set_xlabel('Categories')
axs[1].set_ylabel('Values')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
